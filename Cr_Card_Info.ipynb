```python
#import pandas & matplotlip
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#import csv file & show first couple of rows 
credit_card_data = pd.read_csv('BankChurners.csv')
credit_card_data.tail(20)
```

```python
#make sure there are no empty cells 
credit_card_data.isnull().sum()
```

```python
#describing the data overall
credit_card_data.describe()
```

```python
#describe attrited only
```

```python
#show customers attrited
attrited = credit_card_data.loc[credit_card_data["Attrition_Flag"]== "Attrited Customer"]
attrited.describe()
```

```python
#card category of attrited customers
count_colors = credit_card_data[credit_card_data['Attrition_Flag']== 'Attrited Customer']['Card_Category'].value_counts()
print(count_colors)
```

```python
# %of blue card holders attrited customers 
blue_attrited = (1519/1627)*100
print(blue_attrited)
```

```python
#card category of exsisting customers 
colours_count = credit_card_data[credit_card_data['Attrition_Flag']== 'Existing Customer']['Card_Category'].value_counts()
print(colours_count)
```

```python
# %of blue card holders attrited customers 
blue_existing = (7917/8500)*100
print(blue_existing)
```

```python
#mean credit card limit for attrited customers 
limit = credit_card_data[credit_card_data['Attrition_Flag']== 'Attrited Customer']['Credit_Limit'].mean()
print(limit)
```

```python
#the marital status of attrited customers
Mariage = credit_card_data[credit_card_data['Attrition_Flag']== 'Attrited Customer']['Marital_Status'].value_counts()
print(Mariage)
```

```python
#how many dependants did most attrited customers have
Dependants= credit_card_data[credit_card_data['Attrition_Flag']== 'Attrited Customer']['Dependent_count'].value_counts()
print(Dependants)
```

```python
#for attrited customers , as per their marital status , what was their income category 
credit_card_data[credit_card_data["Attrition_Flag"]== 'Attrited Customer'].groupby('Marital_Status')["Income_Category"].value_counts()
```

```python
#for existing customers , as per their marital status , what was their income category 
credit_card_data[credit_card_data["Attrition_Flag"]== 'Existing Customer'].groupby('Marital_Status')["Income_Category"].value_counts()
```

```python
credit_card_data.groupby('Attrition_Flag')["Income_Category"].value_counts()
```

```python
#Avg age of attrited customers 
Age = credit_card_data[credit_card_data['Attrition_Flag']== 'Attrited Customer']['Customer_Age'].mean()
print(Age)
```

```python
#Avg age of existing customers 
ages = credit_card_data[credit_card_data['Attrition_Flag']== 'Existing Customer']['Customer_Age'].mean()
print(ages)
```

```python
#Attrited customers were mostly male or female
gender = credit_card_data[credit_card_data['Attrition_Flag']== 'Attrited Customer']['Gender'].value_counts()
print(gender)
```

```python
#existing customers are mostly male or female
genders = credit_card_data[credit_card_data['Attrition_Flag']== 'Existing Customer']['Gender'].value_counts()
print(genders)
```
